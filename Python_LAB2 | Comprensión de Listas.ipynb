```python
pares = [i for i in range(1,201) if i%2 == 0]
print(pares)
```

3. Usa una comprensión de lista para crear e imprimir una lista que contenga todos los elementos del siguiente array de 10 x 4.

```python
a = [[0.84062117, 0.48006452, 0.7876326 , 0.77109654],
       [0.44409793, 0.09014516, 0.81835917, 0.87645456],
       [0.7066597 , 0.09610873, 0.41247947, 0.57433389],
       [0.29960807, 0.42315023, 0.34452557, 0.4751035 ],
       [0.17003563, 0.46843998, 0.92796258, 0.69814654],
       [0.41290051, 0.19561071, 0.16284783, 0.97016248],
       [0.71725408, 0.87702738, 0.31244595, 0.76615487],
       [0.20754036, 0.57871812, 0.07214068, 0.40356048],
       [0.12149553, 0.53222417, 0.9976855 , 0.12536346],
       [0.80930099, 0.50962849, 0.94555126, 0.33364763]];
```

```python
elementos = [e for sublista in a for e in sublista]
print(elementos)
```

```python
# Otra forma de hacerlo con el bucle anidado:
elementos = []
for sublista in a:
  for e in sublista:
    elementos.append(e)
print(elementos)
```

4. Añade una condición a la comprensión de lista anterior para que solo se impriman los valores mayores o iguales a 0.5.

```python
elementos_mayores = [e for sublista in a for e in sublista if e >= 0.5]
print(elementos_mayores)
```

```python
# Otra forma de hacerlo con el bucle anidado:
elementos_mayores = []
for sublista in a:
  for e in sublista:
    if e >= 0.5:
      elementos_mayores.append(e)
print(elementos_mayores)
```

5. Usa una comprensión de lista para crear e imprimir una lista que contenga todos los elementos del siguiente array de 5 x 2 x 3.

```python
b = [[[0.55867166, 0.06210792, 0.08147297],
        [0.82579068, 0.91512478, 0.06833034]],

       [[0.05440634, 0.65857693, 0.30296619],
        [0.06769833, 0.96031863, 0.51293743]],

       [[0.09143215, 0.71893382, 0.45850679],
        [0.58256464, 0.59005654, 0.56266457]],

       [[0.71600294, 0.87392666, 0.11434044],
        [0.8694668 , 0.65669313, 0.10708681]],

       [[0.07529684, 0.46470767, 0.47984544],
        [0.65368638, 0.14901286, 0.23760688]]];
```

```python
lista = [e for sublista1 in b for sublista2 in sublista1 for e in sublista2]
print(lista)
```

```python
# Otra forma de hacerlo con el bucle anidado:
lista = []
for sublista1 in b:
  for sublista2 in sublista1:
    for e in sublista2:
      lista.append(e)
print(lista)
```

6. Añade una condición a la comprensión de lista anterior para que se imprima el último valor en cada subarray, pero solo si es menor o igual a 0.5.

```python
lista_menor = [e for sublista1 in b for sublista2 in sublista1 for e in sublista2 if e <= 0.5]
print(lista_menor)
```
