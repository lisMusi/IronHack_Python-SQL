````python
# 1. Filtrar el DataFrame para obtener solo las aplicaciones de pago:
paid_apps = data[data['price'] > 0.00]

# 2. Calcular el precio medio por género para las aplicaciones de pago:
mean_price_by_genre = paid_apps.groupby('prime_genre')['price'].mean()

# 3. Ordenar los géneros por el precio medio de manera descendente:
mean_price_by_genre_sorted = mean_price_by_genre.sort_values(ascending=False)
mean_price_by_genre_sorted
````
