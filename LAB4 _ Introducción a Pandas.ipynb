10. Cree un Pandas DataFrame a partir del diccionario de pedidos de productos que se muestra a continuación.

```python
orders = {'Description': ['LUNCH BAG APPLE DESIGN',
  'SET OF 60 VINTAGE LEAF CAKE CASES ',
  'RIBBON REEL STRIPES DESIGN ',
  'WORLD WAR 2 GLIDERS ASSTD DESIGNS',
  'PLAYING CARDS JUBILEE UNION JACK',
  'POPCORN HOLDER',
  'BOX OF VINTAGE ALPHABET BLOCKS',
  'PARTY BUNTING',
  'JAZZ HEARTS ADDRESS BOOK',
  'SET OF 4 SANTA PLACE SETTINGS'],
 'Quantity': [1, 24, 1, 2880, 2, 7, 1, 4, 10, 48],
 'UnitPrice': [1.65, 0.55, 1.65, 0.18, 1.25, 0.85, 11.95, 4.95, 0.19, 1.25],
 'Revenue': [1.65, 13.2, 1.65, 518.4, 2.5, 5.95, 11.95, 19.8, 1.9, 60.0]}
```

```python
orders_df = pd.DataFrame(orders)
orders_df
```

11. Calcule la cantidad total pedida y los ingresos generados por estos pedidos.

```python
cantidad_total = orders_df['Quantity'].sum()
ingresos_totales = orders_df['Revenue'].sum()

print('La cantidad total pedida es:', cantidad_total)
print('Los ingresos generados por estos pedidos son de', ingresos_totales,'euros')  # suponiendo que la empresa sea europea
```

12. Obtenga los precios de los artículos más caros y menos caros pedidos e imprima la diferencia.

```python
articulo_mas_caro = orders_df['UnitPrice'].max()
articulo_mas_barato = orders_df['UnitPrice'].min()
diferencia = articulo_mas_caro - articulo_mas_barato

print('La diferencia entre el artículo mas caro y el menos caro es de', diferencia, 'euros') # suponiendo que la empresa sea europea
```
